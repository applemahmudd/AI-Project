**TTS**

```python
from gtts import gTTS
from IPython.display import Audio, display

text = "Hi how are you"
lang = "en"

tts = gTTS(text=text, lang=lang, slow=False)
tts.save("audio.mp3")

display(Audio("audio.mp3", autoplay=True))


```
